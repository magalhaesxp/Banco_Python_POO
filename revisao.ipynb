```python
#Importando Biblioteca
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use('seaborn')
```

```python
#criando um dataframe
df = pd.read_excel('AdventureWorks.xlsx')
```

```python
#Visualizando as 5 primeiras linhas
df.head()
```

```python
#Quantidade de linhas e colunas
df.shape
```

```python
#Verificando o tipo de dados
df.dtypes
```

```python
#Valor total da receita
df['Valor Venda'].sum()
```

```python
#Custo total por item
df['custo'] = df['Custo Unitário'].mul(df['Quantidade'])
```

```python
df.head(1)
```

```python
# custo total
round(df['custo'].sum(), 2)
```

```python
#Lucro total
df['lucro'] = df['Valor Venda'] - df['custo']
```

```python
df.head(1)
```

```python
#lucro total
round(df['lucro'].sum(),2)
```

```python
# Criar coluna com total de dadas
df['Tempo_envio'] = df['Data Envio'] - df['Data Venda']
```

```python
df.head(1)
```

```python
#Extraindo os dias
df['Tempo_envio'] = (df['Data Envio'] - df['Data Venda']).dt.days
```

```python
df.head(1)
```

```python
df["Tempo_envio"].dtypes
```

```python
#media de tempo de envio por marca
df.groupby('Marca')['Tempo_envio'].mean()
```

```python
#Verifica se existe campos nulos
df.isnull().sum()
```

```python
#Agrupar por ano e marca
df.groupby([df['Data Venda'].dt.year, 'Marca'])['lucro'].sum().reset_index()
```

```python
# Ajusta todos os formatos flutuantes
pd.options.display.float_format = '{:20,.2f}'.format
```

```python
#Resetando o index
lucro_ano = df.groupby([df['Data Venda'].dt.year, "Marca"])['lucro'].sum().reset_index()
lucro_ano
```

```python
df.groupby('Produto')['Quantidade'].sum().sort_values(ascending=False)
```

```python
#Gráfico Total de produtos vendidos
df.groupby('Produto')['Quantidade'].sum().sort_values(ascending=False).plot.barh(title='Total de produtos vendidos')
plt.xlabel('Total')
plt.ylabel('Produto');
```

```python
#Lucro por ano
df.groupby(df['Data Venda'].dt.year)['lucro'].sum().plot.bar(title='Lucro x Ano')
plt.xlabel('Ano')
plt.ylabel('Receita');
```

```python
df.groupby(df['Data Venda'].dt.year)['lucro'].sum()
```

```python
#selecionando as vendas de 2009
df_2009 = df[df['Data Venda'].dt.year == 2009]
df_2009.head()
```

```python
df_2009.groupby(df['Data Venda'].dt.month)['lucro'].sum().plot(title='Lucro x Mês')
plt.xlabel('Mês')
plt.ylabel('Lucro');
```

```python
df_2009.groupby(df['Marca'])['lucro'].sum().plot.bar(title='Lucro x Marca')
plt.xlabel('Marca')
plt.ylabel('Lucro')
plt.xticks(rotation='horizontal');
```

```python
df_2009.groupby(df['Classe'])['lucro'].sum().plot.bar(title='Lucro x Classe')
plt.xlabel('Classe')
plt.ylabel('Lucro');
```
